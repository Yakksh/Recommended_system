```python
import pandas as pd
pd.set_option('display.max_columns', 300)
pd.set_option('display.max_rows', 300)
pd.set_option('display.width', 1000)
```

```python
movies = pd.read_csv('dataset/movies.csv')
ratings = pd.read_csv('dataset/ratings.csv')
```

```python
movies
```

```python
ratings
```

```python
data = pd.merge(movies, ratings).drop(['genres', 'timestamp'], axis=1)
data
```

```python
user_ratings = data.pivot_table(index='userId', columns='title', values='rating')
user_ratings
```

```python
user_ratings = user_ratings.dropna(thresh=10, axis=1).fillna(0)
user_ratings
```

```python
item_similarity = user_ratings.corr(method='pearson')
item_similarity
```

```python
def get_similar_movies(movie_name, user_rating):
    similarity_score = item_similarity[movie_name]*(user_rating-2.5)
    similarity_score = similarity_score.sort_values(ascending=False)
    return similarity_score
```

```python
user = [('10 Things I Hate About You (1999)',5),
        ('Zootopia (2016)',3),
        ('xXx (2002)',1)]

similar_movies = pd.DataFrame()

for movie,rating in user:
    similar_movies = similar_movies.append(get_similar_movies(movie, rating), ignore_index=True)
    
```
